### –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∏ –¥–æ–±–∞–≤—å—Ç–µ –≤ —Ç–∞–±–ª–∏—Ü—É –Ω–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã

    —Ü–µ–Ω–∞ –∫–≤–∞—Ä–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ - 'sqm_price'

    –¥–µ–Ω—å –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è - 'day_expo'

    –º–µ—Å—è—Ü –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ - 'month_expo'

    –≥–æ–¥ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ - 'year_expo'

    —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –≤ –∫–º –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞ - 'center_distance_km'

    —Ç–∏–ø —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã - 'floor_type' (–ø–µ—Ä–≤—ã–π, –ø–æ—Å–ª–µ–¥–Ω–∏–π, –¥—Ä—É–≥–æ–π) - –Ω–∞–ø–∏—à–µ–º —á–µ—Ä–µ–∑ –æ—Ç–¥–µ–ª—å–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –∏ –º–µ—Ç–æ–¥ .apply()

````python
data['sqm_price'] = data['last_price']/data['total_area']
data['day_expo'] = data['first_day_exposition'].dt.dayofweek # –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫ = 0, –≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ = 6.
data['year_expo'] = data['first_day_exposition'].dt.year
data['month_expo'] = data['first_day_exposition'].dt.month
data['center_distance_km'] = round(data['city_centers_nearest']/1000) #–ø—Ä–∏–≤–æ–¥–∏–º –º–µ—Ç—Ä—ã –∫ –∫–∏–ª–æ–º–µ—Ç—Ä–∞–º –∏ –æ–∫—Ä—É–≥–ª—è–µ–º
````

````python
def level(row):
    all_level = row['floors_total']
    level = row['floor']
    if all_level == level:
        row['floor_type'] = '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
        return row
    elif level == 1:
        row['floor_type'] = '–ø–µ—Ä–≤—ã–π'
        return row
    else:
        row['floor_type'] = '–¥—Ä—É–≥–æ–π'
        return row
    
data = data.apply(level, axis=1)
````
